```python
import pandas as pd
```

```python
df202007 = pd.read_csv("JC-202007-citibike-tripdata.csv")
df202006 = pd.read_csv("JC-202006-citibike-tripdata.csv")
df202005 = pd.read_csv("JC-202005-citibike-tripdata.csv")
df202004 = pd.read_csv("JC-202004-citibike-tripdata.csv")
df202003 = pd.read_csv("JC-202003-citibike-tripdata.csv")
df202002 = pd.read_csv("JC-202002-citibike-tripdata.csv")
df202001 = pd.read_csv("JC-202001-citibike-tripdata.csv")
df201912 = pd.read_csv("JC-201912-citibike-tripdata.csv")
df201911 = pd.read_csv("JC-201911-citibike-tripdata.csv")
df201910 = pd.read_csv("JC-201910-citibike-tripdata.csv")
df201909 = pd.read_csv("JC-201909-citibike-tripdata.csv")
df201908 = pd.read_csv("JC-201908-citibike-tripdata.csv")
df201907 = pd.read_csv("JC-201907-citibike-tripdata.csv")
df202007.head()
```

```python
df = pd.concat([df201907, df201908, df201909, df201910, df201911, df201912, df202001, \
           df202002, df202003, df202004, df202005, df202006, df202007])
df.head()
```

```python
pctg = round((len(df202007) / len(df201907) - 1)*100, 1)
print(f"From 2019-08 to 2020-08, total ridership has grown by {pctg}%.")
```

```python
sub202007 = df202007["usertype"].value_counts()["Subscriber"]
print("2020-07")
print(f"Proportion of subscribers: {sub202007 / df202007['usertype'].count()}")
cus202007 = df202007["usertype"].value_counts()["Customer"]
print(f"Proportion of customers: {cus202007 / df202007['usertype'].count()}")

sub201907 = df201907["usertype"].value_counts()["Subscriber"]
print("---\n2019-07")
print(f"Proportion of subscribers: {sub201907 / df201907['usertype'].count()}")
cus201907 = df201907["usertype"].value_counts()["Customer"]
print(f"Proportion of customers: {cus201907 / df201907['usertype'].count()}")
```

```python
summer = pd.concat([df202006, df202007, df201908, df201907])
summer["hour"] = pd.to_datetime(summer.starttime)
summer["hour"] = summer["hour"].dt.hour
summer.groupby(["hour"])["bikeid"].count()
```

```python
winter = pd.concat([df201912, df202001, df202002])
winter["hour"] = pd.to_datetime(winter.starttime)
winter["hour"] = winter["hour"].dt.hour
winter.groupby(["hour"])["bikeid"].count()
```

```python
d='2019-09-09' 
totay_date=pd.to_datetime(d).date()

df201909["date"] = pd.to_datetime(df201909["starttime"])
df201909["date"] = df201909["date"].dt.date
df_queried_date = df201909.loc[df201909["date"] == totay_date]
```

```python
df_top_stations = df_queried_date.groupby(["start station id"])\
    ["start station id"].count().sort_values(ascending=False)[0:10].to_frame()
df_top_stations = df_top_stations.rename(columns={"start station id":"number of start stations"})
df_top_stations["start station id"] = df_top_stations.index
df_top_stations.reset_index(drop=True, inplace=True)
# df_top_stations
```

```python
# Report top station names
stations = df201909[["start station id", "start station name"]].drop_duplicates()
df_top_stations.merge(stations, how="inner", on="start station id")
```

```python
# Find top stations
df_top_stations = df_queried_date.groupby(["end station id"])\
    ["end station id"].count().sort_values(ascending=False)[0:10].to_frame()
df_top_stations = df_top_stations.rename(columns={"end station id":"number of end stations"})
df_top_stations["end station id"] = df_top_stations.index
df_top_stations.reset_index(drop=True, inplace=True)
# df_top_stations
```

```python
# Report top station names
stations = df201909[["end station id", "end station name"]].drop_duplicates()
df_top_stations.merge(stations, how="inner", on="end station id")
```
